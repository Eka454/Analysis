```python
df.columns = map(str.lower, df.columns)
```

```python
df.info()
```

```python
df['sales_sum'] = df[['na_sales','eu_sales','jp_sales']].sum(axis=1)
max(df[df.user_score == 'tbd'].sales_sum)
```

```python
df.year_of_release = pd.to_numeric(df.year_of_release, errors='coerce').astype('Int64')
df.user_score = df['user_score'].replace('tbd', np.NaN)
df.user_score = df.user_score.astype('float64')
```

```python
names_1 = df[df.rating.isnull()].name.unique()
names_2 = df[~df.rating.isnull()].name.unique()
union_names = names_1[np.in1d(names_1, names_2)]
name_rating_True = df[df['name'].isin(union_names) & ~df.rating.isnull()][['name', 'rating']] \
                                                                        .drop_duplicates(keep=False, inplace=False)
```

```python
def replace(name):
    for stroka in name_rating_True.iloc:
        if name == stroka[0]:
            return stroka[1]
```

```python
replace_rating = df[df['name'].isin(union_names) & df.rating.isnull()].name.apply(replace)
df.loc[replace_rating.keys(), ['rating']] = replace_rating
```

<font color='green'><b>Вывод:</b> тип данных был заменен в столбце year_of_release, так как год всегда подразумевает целое число, и для столбца user_score был установлен тип float, а 'tbd' заменен на nan. Пропуски имеются в столбцах name, year_of_release, genre, critic_score, user_score, rating.  
    Пропуски в рейтингах могут быть за счет минимальной востребованности игр и в результате отсутствия оценки.
    Пропуски были заменены в столбце rating, так как этот столбец определяет возрастной интервал для игроков, то он совпадает для одинаковых игр, запущенных на разных платформах, поэтому rating можно выставить, если name игры уже существует в df с заданным rating. Аббревиатура 'tbd' может означать слишком маленькое количество отзывов, поэтому оценка может быть недостаточно точной, так как оценку 'tbd' имеют игры с максимальными суммарными продажами равными 6,04 млн.

# Шаг 4. Составьте портрет пользователя каждого региона

Определите для пользователя каждого региона (NA, EU, JP):

* Самые популярные платформы (топ-5). Опишите различия в долях продаж.
* Самые популярные жанры (топ-5). Поясните разницу.
* Влияет ли рейтинг ESRB на продажи в отдельном регионе?


```python
def top_5(column):
    na = since_2005.groupby([column])['na_sales'].sum().nlargest(5).sort_values(ascending = False)
    eu = since_2005.groupby([column])['eu_sales'].sum().nlargest(5).sort_values(ascending = False)
    jp = since_2005.groupby([column])['jp_sales'].sum().nlargest(5).sort_values(ascending = False)
    
    fna = na.to_frame()
    fna[column] = fna.index
    fna = fna[[column, 'na_sales']].reset_index(drop=True)

    feu = eu.to_frame()
    feu[column] = feu.index
    feu = feu[[column, 'eu_sales']].reset_index(drop=True)

    fjp = jp.to_frame()
    fjp[column] = fjp.index
    fjp = fjp[[column, 'jp_sales']].reset_index(drop=True) 
    
    return pd.concat([fna, feu, fjp], axis=1)
```

```python
top_platfroms = top_5('platform')
print('Топ-5 платформ для каждого региона')
top_platfroms 
```

```python
top_genre = top_5('genre')
print('Топ-5 жанров для каждого региона')
top_genre 
```

```python
top_ESRB = top_5('rating')
print('Топ-5 востребованных ESRB для каждого региона')
top_ESRB
```

# Шаг 5. Проверьте гипотезы

* Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
* Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.  

Задайте самостоятельно пороговое значение alpha.  
Поясните:
* Как вы сформулировали нулевую и альтернативную гипотезы;
* Какой критерий применили для проверки гипотез и почему.


```python
from math import sqrt
```

Нулевая гипотеза - между двумя наборами данных нет разницы. Средние пользовательские рейтинги XOne и PC одинаковые.  
Альтернативная гипотеза - Средние пользовательские рейтинги XOne и PC разные.  

```python
XOne = since_2005[(since_2005.platform == 'XOne') & ~since_2005.user_score.isnull()]
XOne_i = XOne.user_score
XOne_mean = XOne_i.mean()

PC = since_2005[(since_2005.platform == 'PC') & ~since_2005.user_score.isnull()]
PC_i = PC.user_score
PC_mean = PC_i.mean()

s_XOne = sum((XOne_i-XOne_mean)**2)
s_PC = sum((PC_i-PC_mean)**2)
n1 = len(XOne_i)
n2 = len(PC_i)
           
d = sqrt( ((s_XOne + s_PC) / (n1 + n2 - 2))  \
    * ((n1 + n2)/(n1 * n2)) )
t = (PC_mean - XOne_mean) / d
ss = n1 + n2 - 2

print('Степень свободны: ', ss, '\nt эмпирический: ', t, '\nt критическое при уровне значимости 0.05: 1.9628471')
```

Нулевая гипотеза - Средние пользовательские рейтинги Sports и Action одинаковые.  
Альтернативная гипотеза - Средние пользовательские рейтинги Sports и Action разные.  

```python
Sports = since_2005[(since_2005.genre == 'Sports') & ~since_2005.user_score.isnull()]
Sports_i = Sports.user_score
Sports_mean = Sports_i.mean()

Action = since_2005[(since_2005.genre == 'Action') & ~since_2005.user_score.isnull()]
Action_i = Action.user_score
Action_mean = Action_i.mean()

s_Action = sum((Action_i-Action_mean)**2)
s_Sports = sum((Sports_i-Sports_mean)**2)
n1 = len(Action_i)
n2 = len(Sports_i)
           
d = sqrt( ((s_Action + s_Sports) / (n1 + n2 - 2))  \
    * ((n1 + n2)/(n1 * n2)) )
t = (Action_mean - Sports_mean) / d
ss = n1 + n2 - 2

print('Степень свободны: ', ss, '\nt эмпирический: ', t, '\nt критическое при уровне значимости 0.05: 1.9623391')
```

<font color='green'><b>Вывод:</b>
Стат значимость вычислялась по критерию Стьюдента. Так как t эмпирическая больше t критического, то принимается альтернативная гипотеза. Следовательно разница между XOne, PC, а такжде между Sports и Action не является стат значимой
