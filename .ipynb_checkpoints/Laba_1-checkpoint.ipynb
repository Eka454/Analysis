```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
import matplotlib.ticker as ticker
```

```python
data = pd.read_csv('laba_1.csv')
data.start_session = pd.to_datetime(data['start_session'])
data.end_session = pd.to_datetime(data['end_session'])
```

```python
data.head()
```

```python
duration = data.end_session - data.start_session
hours = duration / np.timedelta64(1, 'h')

def int_r(num):
    num = int(num + (0.5 if num > 0 else -0.5))
    return num

hour = []
for i in range(len(hours)):
    hour += [int_r(hours[i])]
cnt_h = dict(Counter(hour))
```

```python
x = cnt_h.keys()
y = cnt_h.values()

fig, ax = plt.subplots()

ax.bar(x, y)
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))

ax.set_facecolor('seashell')
fig.set_facecolor('floralwhite')
fig.set_figwidth(12)    #  ширина Figure
fig.set_figheight(6)    #  высота Figure
ax.set_xlabel('сессии')
ax.set_ylabel('время (ч)')
ax.set_title('Длина сессий')

plt.show()
```

### 1) В основном сессии длятся около 1-2 часов 

```python
start = pd.to_datetime(data.start_session.astype(str)).dt.hour
end = pd.to_datetime(data.end_session.astype(str)).dt.hour
times = []
for i in range(len(data)):
    times += list(range(start[i], end[i]))
cnt = dict(Counter(times))
```

```python
x = cnt.keys()
y = cnt.values()

fig, ax = plt.subplots()

ax.bar(x, y)
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))

ax.set_facecolor('seashell')
fig.set_facecolor('floralwhite')
fig.set_figwidth(12)    #  ширина Figure
fig.set_figheight(6)    #  высота Figure
ax.set_xlabel('часы')
ax.set_ylabel('посещения')
ax.set_title('Зависимость числа сессий от времени суток')

plt.show()
```
